Task - 3 - Outlier Detection:


Tasks:
1. Identify outlier countries in South Asia based on their Score and GDP per Capita.

```python
# Load the dataset
file_path = '/content/drive/My Drive/Asignment of math/WHR-2024-5CS037.csv'
south_asia_data = pd.read_csv(file_path)


q1_score = south_asia_data_filtered['score'].quantile(0.25)
q3_score = south_asia_data_filtered['score'].quantile(0.75)
iqr_score = q3_score - q1_score

q1_gdp = south_asia_data_filtered['Log GDP per capita'].quantile(0.25)
q3_gdp = south_asia_data_filtered['Log GDP per capita'].quantile(0.75)
iqr_gdp = q3_gdp - q1_gdp

lower_bound_score = q1_score - 1.5 * iqr_score
upper_bound_score = q3_score + 1.5 * iqr_score

lower_bound_gdp = q1_gdp - 1.5 * iqr_gdp
upper_bound_gdp = q3_gdp + 1.5 * iqr_gdp

outliers = south_asia_data_filtered[
    (south_asia_data_filtered['score'] < lower_bound_score) |
    (south_asia_data_filtered['score'] > upper_bound_score) |
    (south_asia_data_filtered['Log GDP per capita'] < lower_bound_gdp) |
    (south_asia_data_filtered['Log GDP per capita'] > upper_bound_gdp)
]

# Display outliers
print("Outlier Countries in South Asia based on Score and GDP per Capita:")
print(outliers[['Country name', 'score', 'Log GDP per capita']])
```

2. Define outliers using the 1.5 Ã— IQR rule.

 => According to 1.5*IQR rule we can say that outliers are data points that are different from the rest of the dataset. 1.5*IQR rule is commont method that is used to detect the ouliers which is based on IQR. The difference between q3 and q1 (IQR=q3-q1)is  determined by lower and upper Bounds.

 We calculate lower and upper bounds:

 i>lower bounds=q1-1.5*IQR

 ii>upper bounds=q3+1.5*IQR

After all we can say that any of the data points which is above the upper bonds or below the lower bonds are called outlier.To identify the value which are far from the majority of data will helps to find from this method.

6. Outlier Detection:


â€¢ Identify outlier countries in both regions based on Score and GDP per Capita.

```python
import pandas as pd

# Load the dataset
file_path = '/content/drive/My Drive/Asignment of math/WHR-2024-5CS037.csv'
data = pd.read_csv(file_path)

# List of countries in South Asia and Middle East
south_asian_countries = [
    "Afghanistan", "Bangladesh", "Bhutan", "India", "Maldives", "Nepal", "Pakistan", "Sri Lanka"
]
middle_east_countries = [
    "Bahrain", "Iran", "Iraq", "Israel", "Jordan", "Kuwait", "Lebanon", "Oman",
    "Palestine", "Qatar", "Saudi Arabia", "Syria", "United Arab Emirates", "Yemen"
]

south_asia_data = data[data['Country name'].isin(south_asian_countries)]
middle_east_data = data[data['Country name'].isin(middle_east_countries)]

def identify_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

south_asia_score_outliers = identify_outliers(south_asia_data, 'score')
south_asia_gdp_outliers = identify_outliers(south_asia_data, 'Log GDP per capita')

print("\nSouth Asia: Outlier countries based on Score:")
print(south_asia_score_outliers[['Country name', 'score']])

print("\nSouth Asia: Outlier countries based on GDP per Capita:")
print(south_asia_gdp_outliers[['Country name', 'Log GDP per capita']])

middle_east_score_outliers = identify_outliers(middle_east_data, 'score')
middle_east_gdp_outliers = identify_outliers(middle_east_data, 'Log GDP per capita')

# Print outlier countries for Middle East
print("\nMiddle East: Outlier countries based on Score:")
print(middle_east_score_outliers[['Country name', 'score']])

print("\nMiddle East: Outlier countries based on GDP per Capita:")
print(middle_east_gdp_outliers[['Country name', 'Log GDP per capita']])
```

â€¢ Plot these outliers and discuss their implications.

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

def plot_outliers(data, outliers, region, title):
    plt.figure(figsize=(10, 6))

    # Plot all data points
    sns.scatterplot(
        data=data,
        x='Log GDP per capita',
        y='score',
        color='blue',
        label='Normal',
        s=100, alpha=0.6
    )

    # Highlight outliers in red
    sns.scatterplot(
        data=outliers,
        x='Log GDP per capita',
        y='score',
        color='red',
        label='Outliers',
        s=200, edgecolor='black', marker='X'
    )

    # Set plot details
    plt.title(f'{region}: Outliers Based on Score and GDP per Capita', fontsize=14)
    plt.xlabel('Log GDP per Capita', fontsize=12)
    plt.ylabel('Happiness Score', fontsize=12)
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

if not south_asia_score_outliers.empty or not south_asia_gdp_outliers.empty:
    plot_outliers(south_asia_data, south_asia_score_outliers, 'South Asia', 'Outliers in South Asia')
else:
    print("No outliers found in South Asia.")

if not middle_east_score_outliers.empty or not middle_east_gdp_outliers.empty:
    plot_outliers(middle_east_data, middle_east_score_outliers, 'Middle East', 'Outliers in Middle East')
else:
    print("No outliers found in Middle East.")
```
